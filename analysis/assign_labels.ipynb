```python
import pandas as pd
```

```python
# Get the mappings from channel id to political label, turn into a dictionary
# Copied from old analysis code (analysis/assign_labels.ipynb)
hosseinmardi_channels_df = pd.read_csv('../labels/h_mbfc_labels.csv')
h_id_channels = dict(zip(
    hosseinmardi_channels_df['CHANNEL_ID'],
    hosseinmardi_channels_df['H_TAG']
))
h_name_channels = dict(zip(
    hosseinmardi_channels_df['CHANNEL_TITLE'],
    hosseinmardi_channels_df['H_TAG']
))
h_id_channels.update(h_name_channels)
h_channels = h_id_channels
```

```python
len(h_channels.keys())
```

```python
# Put political labels on each video in the results
# Modified from old analysis code (analysis/assign_labels.ipynb)
# No name checking
def get_political_label(channel_id, channel_labels):
    # Find the channel's corresponding labels
    if channel_id in channel_labels.keys():
        return channel_labels[channel_id]
#    if channel_name in channel_labels.keys():
#        return channel_labels[channel_name]
    return 'Other'
```

```python
results_filepath_prefix = '../results/'
result_filename = 'test_channels.csv'
result_filepath = results_filepath_prefix + result_filename

result_df = pd.read_csv(result_filepath)
result_df.head()
```

```python
labels = result_df.apply(
    lambda row: get_political_label(row['channel_id'],
                                    h_channels),
    axis = 1
)
result_df['political_label'] = labels
result_df.head()
```

```python
result_df.to_csv(result_filepath)
```
